```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline

import os
import numpy as np
import pandas as pd
from pathlib import Path

from crop_resize import Resize
```

```python
height = 137
width = 236
image_size = 128
data_type = 'train'
in_dir = Path('../input/bengaliai-cv19')
out_dir = Path('../input/bengaliai-cv19-feather')
if not os.path.isdir(outdir): os.mkdir(outdir)
```

```python
for i in range(4):
    df = pd.read_parquet(in_dir/f'{data_type}_image_data_{i}.parquet')
    df = Resize(df, width, height, image_size)
    df.to_feather(out_dir/f'{data_type}_image_data_1x{image_size}x{image_size}_{i}.feather')
```

```python
from crop_resize import read_feathers

filenames = [out_dir/f'{data_type}_image_data_1x{image_size}x{image_size}_{i}.feather' for i in range(4)]
images = read_feathers(filenames, image_size)
images.shape
```

```python
import matplotlib.pyplot as plt

for i in range(10):
    plt.imshow(images[i], cmap=plt.cm.binary)
    plt.show()
```
